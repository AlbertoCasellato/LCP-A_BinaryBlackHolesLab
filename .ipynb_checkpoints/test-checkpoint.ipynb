```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import sys
sys.path.append("src")

from data_preprocessing import load_data
```

```python
data = load_data("data")
data["A1"]["0.0002"]
```

```python
def plot_histogram(data_1, data_2, metallicities, bins=100, xlim=None, xlabel="", title=""):
    """
    Plots histograms for two sets of data with the same bins and xlim.
    """
    # plot settings
    major_ticks_length = 10
    minor_ticks_length = major_ticks_length / 2
    label_size = 14
    
    fig, axes = plt.subplots(2, 1, figsize=(8, 8), sharex=True, gridspec_kw={"hspace": 0})
    cmap = cm.plasma
    colors = [cmap(i / len(data_1)) for i in range(len(data_1))]
    if xlim:
        bins = np.linspace(xlim[0], xlim[1], bins)
    
    # plots
    for i in range(len(data_1)):
        axes[0].hist(data_1[i], bins=bins, histtype="step", label=f"Z={metallicities[i]}", edgecolor=colors[i], linewidth=1.5)
        axes[1].hist(data_2[i], bins=bins, histtype="step", label=f"Z={metallicities[i]}", edgecolor=colors[i], linewidth=1.5)
    
    # axes settings
    axes[0].legend(ncol=2, fontsize="small", loc="upper right")
    axes[0].tick_params(axis='both', which='both', direction="in", top=True, right=True, labelbottom=False)
    axes[1].tick_params(axis='both', which='both', direction="in", top=True, right=True, labelbottom=True)
    
    axes[0].set_ylabel("$N$ Systems\n (Stable Mass Transfer)", fontsize=label_size)
    axes[1].set_ylabel("$N$ Systems\n (Common Envelope)", fontsize=label_size)
    axes[1].set_xlabel(xlabel, fontsize=label_size)

    for ax in axes:
        if xlim:
            ax.set_xlim(xlim)
        ax.set_yscale("log")
        ax.tick_params(which="major", length=major_ticks_length, labelsize=label_size-2)
        ax.tick_params(which="minor", length=minor_ticks_length)

    fig.suptitle(title, fontsize=label_size+2)
    plt.tight_layout()
    plt.show()

def plot_stackplot(metallicities, smt_fractions, ce_fractions, alpha):
    """
    Plots a stackplot showing fractions of stable mass transfer and common envelope systems.
    """
    plt.figure(figsize=(8, 6))
    plt.stackplot(
        metallicities,
        smt_fractions,
        ce_fractions,
        labels=["Stable Mass Transfer", "Common Envelope"],
        colors=["#1f77b4", "#ff7f0e"],
        edgecolor="k",
        alpha=0.8
    )
    plt.xscale("log")
    plt.xlabel("Metallicity")
    plt.ylabel("Fraction")
    plt.title(f"CE vs Stable Mass Transfer Fractions (\u03B1 = {alpha})")
    plt.legend(loc="upper left")
    plt.tight_layout()
    plt.show()
```

```python
# TODO: check from task_1.ipynb what changed
alpha_keys = sorted(data.keys(), key=lambda x: float(x[1:]))
for alpha_key in alpha_keys:
    alpha = float(alpha_key[1:])
    alpha_data = data[alpha_key]

    smt_fractions = []
    ce_fractions = []

    hist_data_smt = []
    hist_data_ce = []
    bh_mass_smt = []
    bh_mass_ce = []
    delay_times_smt = []
    delay_times_ce = []
    sma_smt = []
    sma_ce = []
    ecc_smt = []
    ecc_ce = []

    metallicities = sorted(alpha_data.keys(), key=lambda x: float(x))
    for metallicity in metallicities:
        df = alpha_data[metallicity]

        df_smt = df[df["col.21:CE"] == 0]
        df_ce = df[df["col.21:CE"] == 1]

        total_systems = len(df)
        smt_fractions.append(len(df_smt) / total_systems)
        ce_fractions.append(len(df_ce) / total_systems)

        hist_data_smt.append(df_smt["col.1:m1ZAMS/Msun"].values)
        hist_data_ce.append(df_ce["col.1:m1ZAMS/Msun"].values)

        bh_mass_smt.append(df_smt["col.3:m1rem/Msun"].values)
        bh_mass_ce.append(df_ce["col.3:m1rem/Msun"].values)

        delay_times_smt.append(df_smt["col.6:delay_time/Myr"].values)
        delay_times_ce.append(df_ce["col.6:delay_time/Myr"].values)

        sma_smt.append(df_smt["col.7:sma/Rsun"].values)
        sma_ce.append(df_ce["col.7:sma/Rsun"].values)

        ecc_smt.append(df_smt["col.8:ecc"].values)
        ecc_ce.append(df_ce["col.8:ecc"].values)

    plot_histogram(
        hist_data_smt, hist_data_ce, metallicities,
        xlabel="Primary Mass [M$_\odot$]",
        title=f"Primary Mass Distribution (α = {alpha})"
    )

    plot_histogram(
        bh_mass_smt, bh_mass_ce, metallicities,
        xlabel="Black Hole Mass [M$_\odot$]",
        title=f"Black Hole Mass Distribution (α = {alpha})"
    )

    plot_histogram(
        delay_times_smt, delay_times_ce, metallicities,
        xlabel="Delay Time [Myr]",
        title=f"Delay Time Distribution (α = {alpha})"
    )

    plot_histogram(
        sma_smt, sma_ce, metallicities,
        xlabel="Semi-Major Axis [R$_\odot$]",
        title=f"Semi-Major Axis Distribution (α = {alpha})"
    )

    plot_histogram(
        ecc_smt, ecc_ce, metallicities,
        xlabel="Eccentricity",
        title=f"Eccentricity Distribution (α = {alpha})"
    )

    plot_stackplot(
        [float(z) for z in metallicities],
        smt_fractions, ce_fractions, alpha
    )

    print(f"Completed analysis for α = {alpha}")
    print("-" * 70)
    break
```
